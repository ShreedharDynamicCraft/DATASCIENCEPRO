```python
from flask import Flask, render_template
from bs4 import BeautifulSoup
import requests

app = Flask(__name__)

# Route for the home page
@app.route('/')
def index():
    return render_template('index.html')

# Route for scraping YouTube data
@app.route('/youtube')
def youtube():
    youtube_data = scrape_youtube()
    return render_template('youtube.html', data=youtube_data)

# Route for scraping Amazon data
@app.route('/amazon')
def amazon():
    amazon_data = scrape_amazon()
    return render_template('index.html', data=amazon_data)

# Function to scrape YouTube data
def scrape_youtube():
    # Example URL for YouTube trending videos
    url = 'https://www.youtube.com/feed/trending'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    videos = []
    # Extracting video titles from the HTML
    for video in soup.find_all('h3', class_='yt-lockup-title'):
        title = video.a.text.strip()
        videos.append(title)
    return videos

# Function to scrape Amazon data
def scrape_amazon():
    # Example URL for Amazon best sellers
    url = 'https://www.amazon.com/best-sellers/zgbs'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    products = []
    # Extracting product titles from the HTML
    for product in soup.find_all('div', class_='p13n-sc-truncated'):
        title = product.a.text.strip()
        products.append(title)
    return products

if __name__ == '__main__':
    app.run(host="0.0.0.0", port=5002)
```

```python
from flask import Flask, render_template_string, request

# Import your recommendation model here
# Since we're providing a placeholder recommendation function, we don't need to import anything
# from recommendation_model import generate_recommendations

app = Flask(__name__)

# Placeholder recommendation function
def generate_recommendations(user_preferences):
    # Placeholder recommendations for different combinations of user preferences
    recommendations = {
        ('Action',): ["Action Movie 1", "Action Movie 2", "Action Movie 3"],
        ('Comedy',): ["Comedy Movie 1", "Comedy Movie 2", "Comedy Movie 3"],
        ('Drama',): ["Drama Movie 1", "Drama Movie 2", "Drama Movie 3"],
        ('Action', 'Comedy'): ["Action-Comedy Movie 1", "Action-Comedy Movie 2", "Action-Comedy Movie 3"],
        ('Action', 'Drama'): ["Action-Drama Movie 1", "Action-Drama Movie 2", "Action-Drama Movie 3"],
        ('Comedy', 'Drama'): ["Comedy-Drama Movie 1", "Comedy-Drama Movie 2", "Comedy-Drama Movie 3"],
        ('Action', 'Comedy', 'Drama'): ["Action-Comedy-Drama Movie 1", "Action-Comedy-Drama Movie 2", "Action-Comedy-Drama Movie 3"]
    }
    
    # Return recommendations based on user preferences
    return recommendations.get(tuple(user_preferences), ["No recommendations available for selected preferences"])


# Homepage route
@app.route('/')
def index():
    return render_template_string("""
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Content Recommendation System</title>
    </head>
    <body>
        <h1>Content Recommendation System</h1>
        <form action="/recommendations" method="post">
            <label for="preferences">Select your preferences:</label><br>
            <input type="checkbox" id="preferences" name="preferences" value="Action">
            <label for="preferences">Action</label><br>
            <input type="checkbox" id="preferences" name="preferences" value="Comedy">
            <label for="preferences">Comedy</label><br>
            <input type="checkbox" id="preferences" name="preferences" value="Drama">
            <label for="preferences">Drama</label><br>
            <input type="submit" value="Get Recommendations">
        </form>
    </body>
    </html>
    """)

# Recommendation route
@app.route('/recommendations', methods=['POST'])
def recommendations():
    # Get user preferences from the form
    user_preferences = request.form.getlist('preferences')
    
    # Generate recommendations based on user preferences
    recommended_content = generate_recommendations(user_preferences)
    
    # Render template with recommended content
    return render_template_string("""
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Recommended Content</title>
    </head>
    <body>
        <h1>Recommended Content</h1>
        <ul>
            {% for content in recommendations %}
                <li>{{ content }}</li>
            {% endfor %}
        </ul>
    </body>
    </html>
    """, recommendations=recommended_content)

if __name__ == '__main__':
    app.run(host="0.0.0.0", port=5006)
```
